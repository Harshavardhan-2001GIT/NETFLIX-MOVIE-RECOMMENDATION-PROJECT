```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv('Netflix Movie Recommendation.csv',header = None)
transactions=[]
for i in range(0, 7466):
  transactions.append([str(dataset.values[i,j]) for j in range(0,20)])
```

```python
from apyori import apriori
rules=apriori(transactions=transactions,min_support=0.002,min_confidence=0.2,min_lift=3,min_length=2)
```

```python
results=list(rules)
```

```python
results
```

```python
def inspect(results):
  movie_1=[tuple(result[2][0][0])[0] for result in results]
  movie_2=[tuple(result[2][0][1])[0] for result in results]
  supports= [result[1] for result in results]
  return list(zip(movie_1,movie_2,supports))
resultsinDataFrame = pd.DataFrame(inspect(results),columns=['movie_1','movie_2','supports'])  
```

```python
resultsinDataFrame
```

```python
resultsinDataFrame.nlargest(n=10,columns='supports')
```
